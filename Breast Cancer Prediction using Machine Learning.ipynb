```python
import pandas as pd 
import matplotlib.pyplot as plt       
import seaborn as sns    

from sklearn.pipeline import Pipeline 
from sklearn.model_selection import train_test_split, StratifiedKFold ,GridSearchCV 
from sklearn.decomposition import PCA 
from sklearn.linear_model import LogisticRegression 
from sklearn.preprocessing import StandardScaler 
from sklearn.svm import SVC 
from sklearn.metrics import accuracy_score ,classification_report ,confusion_matrix 
```

```python
X = df.drop(columns=["diagnosis"])
y = df["diagnosis"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=RANDOM_STATE
)
```

```python
print("Dataset size", X.shape)
print("Train Dataset size", X_train.shape)
print("Test Dataset size", X_test.shape)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_scaled, y_train)
```

```python
model.score(X_train_scaled, y_train)
```

```python
model.score(X_test_scaled, y_test)
```

```python
pipeline = Pipeline(
    [
        ("scaler", StandardScaler()),
        ("pca", PCA()),
        ("model", SVC(kernel="rbf"))
    ]
)
```

```python
param_grid = [
    {
        "pca__n_components": [10, 15, 20, 25, 30],
        "model__C": [0.1, 1, 10, 100, 1000],
        "model__gamma": [0.001, 0.01, 0.1, 1, "scale"]
    }
]
```

```python
k=5
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=RANDOM_STATE)
```

```python
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring="recall",
    cv=skf,
    n_jobs=-1
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Hyperparameter Tuning result:")
print("Best params:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

```python
best_pipeline = Pipeline(
    [
        ("scaler", StandardScaler()),
        ("pca", PCA(n_components=10)),
        ("model", SVC(kernel="rbf", C=10, gamma="scale"))
    ]
)
```

```python
best_pipeline.fit(X_train, y_train)
```

```python
y_train_pred = best_pipeline.predict(X_train)
```

```python
train_acc = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {round(train_acc*100, 2)} %")
```

```python

cm = confusion_matrix(y_train, y_train_pred)

plt.figure(figsize=(4,3))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Pred Benign", "Pred Malignant"],
    yticklabels=["Actual Benign", "Actual Malignant"]
)
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix Heatmap")
plt.show()
```

```python
print("Training Classification report")
print(classification_report(y_train, y_train_pred))
```

```python
y_test_pred = best_pipeline.predict(X_test)
```

```python
test_acc = accuracy_score(y_test, y_test_pred)
print(f"Test Accuracy: {round(test_acc*100, 2)} %")
```

```python

cm = confusion_matrix(y_test, y_test_pred)

plt.figure(figsize=(4,3))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Pred Benign", "Pred Malignant"],
    yticklabels=["Actual Benign", "Actual Malignant"]
)

plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix Heatmap")
plt.show()
```

```python
print("Test Classification report")
print(classification_report(y_test, y_test_pred))
```

```python
def predict_cancer(input_features):
    input_df = pd.DataFrame(
        [input_features],
        columns=X_train.columns
    )
    prediction = best_pipeline.predict(input_df)
    print("Model prediction:", prediction)
    if prediction[0] == 1:
        print("Diagnosis - Malignant 🔴")
    else:
        print("Diagnosis - Benign 🟢")
```
